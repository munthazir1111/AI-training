```python
#Binomial Distribution - nc^xp^xq^(n-x)
'''
n,p,q,x
'''
from scipy.stats  import binom
import numpy as np
n = 10
p = 0.5 #q=1-p
x = np.arange(0,n+1)
pmf = binom.pmf(n,p,x)
plt.scatter(x,pmf,color="black")
plt.xlabel("The Pmf Value is")
plt.ylabel("Tha Outcome is")
plt.title("Binomial Distibution ")
plt.show()
```

```python
#Poission Distribution

# calculate the average possible outcomes or rates the 
# interval within the intervel of time e^lambda.lambda^k/n]

from scipy.stats import poisson
lambda_value = 5
k = np.arange(0,10)
pmf = poisson.pmf(lambda_value,k)
print(pmf)
plt.bar(k,pmf,color="yellow")
plt.show()
```

```python
def mean_variance(x, prob_x):
    E = sum(v * p for v, p in zip(x, prob_x))  # Mean
    E2 = sum((v**2) * p for v, p in zip(x, prob_x))  # E(X^2)
    # for  v,p in zip(x,prob_x):
        #  s1 = sum)(v)*p
    #for v,p in zip(x,prob_x):
    # s2=sum(v**2)*p 
    var_value = E2 - E**2  # Variance
    return E, var_value


x = [0, 1, 2]
prob_x = [0.3, 0.5, 0.2]

E, V = mean_variance(x, prob_x)
print(f"The Expected Value (Mean) is: {E}")
print(f"The Variance is: {V}")
```

```python
# central Limit Theoram


import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a non-normal population (e.g., exponential distribution)
population = np.random.exponential(scale=2, size=100000)

# Step 2: Define CLT simulation function
def simulate_clt(population, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))
    return sample_means

# Step 3: Simulate for sample size = 30
sample_means = simulate_clt(population, sample_size=30, num_samples=1000)

# Step 4: Plot results
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='teal', edgecolor='black')
plt.title('Central Limit Theorem Demonstration\nDistribution of Sample Means (n=30)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_dice = 30           # number of dice rolled per trial
n_trials = 1000       # number of trials (samples)
sample_means = []     # to store the mean of each trial

# Simulation
for i in range(n_trials):
    rolls = np.random.randint(1, 7, n_dice)  # roll 30 dice
    sample_means.append(np.mean(rolls))      # compute mean of each trial

# Plot
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black', density=True)
plt.title('Central Limit Theorem with Dice Rolls')
plt.xlabel('Sample Mean of 30 Rolls')
plt.ylabel('Density')
plt.grid(True)
plt.axvline(x=3.5, color='red', linestyle='--', label='Expected Mean (3.5)')
plt.legend()
plt.show()
```
