```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your Excel file
df = pd.read_excel("weather_event_predictor_209.xlsx")

# Conditional Probability Table
table = pd.crosstab(df['WEATHER TYPE'], df['EVENT'], normalize='index').round(2)
print("Conditional Probability (P(Event | Weather Type)):")
print(table)

# Bayes Theorem: P(Sunny | Picnic)
p_sunny = (df['WEATHER TYPE'] == 'Sunny').mean()
p_picnic = (df['EVENT'] == 'Picnic').mean()
p_picnic_given_sunny = table.loc['Sunny', 'Picnic']
result = (p_picnic_given_sunny * p_sunny) / p_picnic

print("\nBayes Theorem: P(Sunny | Picnic)")
print(f"= ({p_picnic_given_sunny} ร {round(p_sunny, 2)}) รท {round(p_picnic, 2)} = {round(result, 2)}")

# Plot 1: Weather Type Count
df['WEATHER TYPE'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Weather Type Frequency")
plt.xlabel("Type")
plt.ylabel("Days")
plt.tight_layout()
plt.show()

# Plot 2: Picnic vs No Picnic (Stacked)
pd.crosstab(df['WEATHER TYPE'], df['EVENT']).plot(kind='bar', stacked=True)
plt.title("Event Count by Weather")
plt.xlabel("Weather Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
